# Truning model

```python
y_train = np.array(y_train)
y_train
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5, random_state = None)

max_k = 30

train_accuracies = [[] for _ in range(max_k)]
val_accuracies = [[] for _ in range(max_k)]

for k in range(max_k):
    knn = KNeighborsClassifier(n_neighbors = k+1)
    
    for train_index, val_index in kf.split(X_train):
        Xtrain, Xval = X_train[train_index], X_train[val_index]
        ytrain, yval = y_train[train_index], y_train[val_index]
        
        knn.fit(Xtrain, ytrain)
        
        
        pred_train = knn.predict(Xtrain)
        pred_val = knn.predict(Xval)
        
        train_accuracies[k].append(accuracy_score(pred_train, ytrain))#each split
        val_accuracies[k].append(accuracy_score(pred_val, yval))

train_accuracies_mean = np.mean(train_accuracies, axis = 1)#each k 
val_accuracies_mean = np.mean(val_accuracies, axis = 1)
```

```python
plt.plot(np.arange(1,max_k+1), train_accuracies_mean, label = "Training Acuuracy")
plt.plot(np.arange(1,max_k+1), val_accuracies_mean, label = "Validation Acuuracy")
plt.legend()
plt.xlabel("k values")
```

```python
def compute_bce_cost(Y, P_hat):

    m = Y.shape[1]  # m -> number of examples in the batch

    cost = (1/m) * np.sum(-Y*np.log(P_hat) - (1-Y)*np.log(1-P_hat))
    cost = np.squeeze(cost)  # remove extraneous dimensions to give just a scalar (e.g. this turns [[17]] into 17)

    dP_hat = (1/m) * (-(Y/P_hat) + ((1-Y)/(1-P_hat)))

    return cost, dP_hat
```
